Dane:

```python
import numpy as np
import pandas as pd

size = 300
X = np.random.rand(size)*5-2.5
w4, w3, w2, w1, w0 = 1, 2, 1, -4, 2
y = w4*(X**4) + w3*(X**3) + w2*(X**2) + w1*X + w0 + np.random.randn(size)*8-4
df = pd.DataFrame({'x': X, 'y': y})
df.to_csv('dane_do_regresji.csv',index=None)
df.plot.scatter(x='x',y='y')
```

Dzielenie zbioru:

```python
from sklearn.model_selection import train_test_split
import numpy as np

print(X.shape, y.shape)

train_x, test_x, train_y, test_y = train_test_split(X, y, train_size=0.8)
print(train_x.shape)
print(test_x.shape)
print(train_y.shape)
print(test_y.shape)

train_x = train_x.reshape(-1,1)
test_x = test_x.reshape(-1,1)
train_y = train_y.reshape(-1,1)
test_x = test_x.reshape(-1,1)
```

Regresja liniowa:

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

linear_reg = LinearRegression()
linear_reg.fit(train_x, train_y)

predicted_y = linear_reg.predict(test_x)

plt.scatter(X, y, color="blue", label="Punkty zbioru")
plt.plot(test_x, predicted_y, color="red", linewidth=2, label="Regresja")

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Regresja liniowa")
plt.show()
```

Regresja wielomianowa:

```python
from sklearn.preprocessing import PolynomialFeatures

poly_f_square = PolynomialFeatures(degree=2, interaction_only=False)

train_x_poly_square = poly_f_square.fit_transform(train_x)
test_x_poly_square = poly_f_square.transform(test_x)

linear_reg_square = LinearRegression()
linear_reg_square.fit(train_x_poly_square, train_y)

print("Intercept:", linear_reg_square.intercept_)
print("Coefficients:", linear_reg_square.coef_)

square_predicted_y = linear_reg_square.predict(test_x_poly_square)

plt.scatter(X, y, color="blue", label="Punkty zbioru")

sorted_indices = np.argsort(test_x.flatten())
test_x_sorted = test_x[sorted_indices]

square_predicted_y_sorted = square_predicted_y[sorted_indices]

plt.plot(test_x_sorted, square_predicted_y_sorted, color="red", linewidth=2, label="Regresja") 

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Regresja kwadratowa")
plt.show()
```

```python
poly_f_3 = PolynomialFeatures(degree=3, interaction_only=False)

train_x_poly_3 = poly_f_3.fit_transform(train_x)
test_x_poly_3 = poly_f_3.transform(test_x)

linear_reg_3 = LinearRegression()
linear_reg_3.fit(train_x_poly_3, train_y)

print("Intercept" , linear_reg_3.intercept_)
print("Coefficients", linear_reg_3.coef_)

predicted_y_3 = linear_reg_3.predict(test_x_poly_3)

plt.scatter(X, y, color="blue", label="Punkty zbioru")

predicted_y_3 = predicted_y_3[ sorted_indices ]
plt.plot(test_x_sorted, square_predicted_y_sorted, color="red", linewidth=2, label="Regresja") 

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Regresja 3-go stopnia")
plt.show()
```

KNN:

```python
from sklearn.neighbors import KNeighborsRegressor

knn_3 = KNeighborsRegressor(n_neighbors=3)
knn_3.fit(train_x.reshape(-1, 1), train_y.reshape(-1, 1))

predicted_knn_3 = knn_3.predict(test_x.reshape(-1, 1))

sorted_indices = np.argsort(test_x.flatten())  
test_x_sorted = test_x[sorted_indices]  
predicted_knn_3_sorted = predicted_knn_3[sorted_indices]  

plt.scatter(X, y, color="blue", label="Punkty zbioru")
plt.plot(test_x_sorted, predicted_knn_3_sorted, color="red", label="Predykcja")

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Regresja k-NN (k=3)")
plt.show()
```

KNN n=5:

```python
from sklearn.neighbors import KNeighborsRegressor

knn_5 = KNeighborsRegressor(n_neighbors=5)
knn_5.fit(train_x.reshape(-1, 1), train_y.reshape(-1, 1))

predicted_knn_5 = knn_5.predict(test_x.reshape(-1, 1))

predicted_knn_5_sorted = predicted_knn_5[sorted_indices]  

plt.scatter(X, y, color="blue", label="Punkty zbioru")
plt.plot(test_x_sorted, predicted_knn_5_sorted, color="red", label="Predykcja")

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Regresja k-NN (k=5)")
plt.show()
```
